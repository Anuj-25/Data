```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Salary_Dataset.csv')
```

```python
X=dataset.iloc[:, 0:-1].values
y=dataset.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.3, random_state=41)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
```

```python
plt.scatter(X_train,y_train,color='red',label='Actual output')
plt.plot(X_train, regressor.predict(X_train),color='blue',label='Predicted output')
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
plt.scatter(X_test,y_test,color='red',label='Actual output')
plt.plot(X_test, regressor.predict(X_test),color='blue',label='Model Prediction')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```
